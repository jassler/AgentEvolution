```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
```

```python
def ternary_plot(data):
    labels = ('R', 'P', 'S')
    sides = 3
    fig_args = {'figsize':(16,16),'facecolor':'white','edgecolor':'white'}
    edge_args = {'color':'black','linewidth':2}
    rotate_labels = False
    label_offset = 0.10
    
    basis = array([
        [
            cos(2*x*pi/sides + start_angle*pi/180),
            sin(2*x*pi/sides + start_angle*pi/180)
        ] for x in range(sides)
    ])
    
    newdata = np.dot(data, basis)
    fig = plt.figure(**figargs)
    ax = fig.add_subplot(111)
    
    for i,l in enumerate(labels):
        if i >= sides:
            break
        
        x = basis[i,0]
        y = basis[i,1]
        if rotate_labels:
            angle = 180 * arctan(y/x)/pi + 90
            if angle > 90 and angle <= 270:
                angle = np.mod(angle + 180, 360)
        else:
            angle = 0
        
        ax.text(
            x * (1 + label_offset),
            y * (1 + label_offset),
            l,
            horizontalalignment='center',
            verticalalignment='center',
            rotation=angle,
            size=34
        )
        
    # clear normal axes
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_frame_on(False)
    
    # plot border
    ax.plot(
        [ basis[x,0] for x in list(range(sides)) + [0,] ],
        [ basis[x,1] for x in list(range(sides)) + [0,] ],
        **edge_args
    )
    
    # extra room for labels
    ax.set_xlim(left=-1.2, right=1.2)
    ax.set_ylim(bottom=-1.2, top=1.2)
    
    # plot
    ax.plot(newdata[:,0], newdata[:,1], linewidth=2, alpha=0.99, c='gray')
    
    fig.show()
```

```python
def plot_file(filename, start=0, end=-1, sep=';', size=(18,5)):
    df = pd.read_csv(filename, sep=sep)
    
    if end < start:
        end = len(df)
    
    if start > 0 or end < len(df):
        df = df[start:end]
    
    df_avg = df[['Avg Rock', 'Avg Paper', 'Avg Scissor']]
    df_lod = df[['LOD Rock', 'LOD Paper', 'LOD Scissor']]
    df_best = df[['Best Rock', 'Best Paper', 'Best Scissor']]
    
    ternary_plot(df_avg)
    ternary_plot(df_lod)
    ternary_plot(df_best)
    
#     # average
#     plt.figure(figsize=size)
#     plt.plot(df['Generation'], df['Avg Rock'], color='grey')
#     plt.plot(df['Generation'], df['Avg Paper'], color='green')
#     plt.plot(df['Generation'], df['Avg Scissor'], color='orange')

#     plt.xlabel('Generation')
#     plt.ylabel('Probability')
#     plt.title('Average Strategy')
#     plt.show()

#     # best
#     plt.figure(figsize=size)
#     plt.plot(df['Generation'], df['Best Rock'], color='grey')
#     plt.plot(df['Generation'], df['Best Paper'], color='green')
#     plt.plot(df['Generation'], df['Best Scissor'], color='orange')

#     plt.xlabel('Generation')
#     plt.ylabel('Probability')
#     plt.title('Best Agent\'s Strategy')
#     plt.show()

#     # line of descent
#     plt.figure(figsize=size)
#     plt.plot(df['Generation'], df['LOD Rock'], color='grey')
#     plt.plot(df['Generation'], df['LOD Paper'], color='green')
#     plt.plot(df['Generation'], df['LOD Scissor'], color='orange')

#     plt.xlabel('Generation')
#     plt.ylabel('Probability')
#     plt.title('Line Of Descent')
#     plt.show()
```
