```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from IPython.core.display import display, HTML
```

```python
df = pd.read_csv('result.csv', sep=';')
df.head()
```

```python
size = (18,5)

# average
plt.figure(figsize=size)
plt.plot(df['Generation'], df['Avg Rock'], color='blue')
plt.plot(df['Generation'], df['Avg Paper'], color='green')
plt.plot(df['Generation'], df['Avg Scissor'], color='orange')

plt.xlabel('Generation')
plt.ylabel('Probability')
plt.title('Average Strategy')
plt.show()

# best
plt.figure(figsize=size)
plt.plot(df['Generation'], df['Best Rock'], color='blue')
plt.plot(df['Generation'], df['Best Paper'], color='green')
plt.plot(df['Generation'], df['Best Scissor'], color='orange')

plt.xlabel('Generation')
plt.ylabel('Probability')
plt.title('Best Agent\'s Strategy')
plt.show()

# line of descent
plt.figure(figsize=size)
plt.plot(df['Generation'], df['LOD Rock'], color='blue')
plt.plot(df['Generation'], df['LOD Paper'], color='green')
plt.plot(df['Generation'], df['LOD Scissor'], color='orange')

plt.xlabel('Generation')
plt.ylabel('Probability')
plt.title('Line Of Descent')
plt.show()
```
